```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
rcParams['figure.figsize'] = [14, 10]
rcParams['font.size'] = 30
```

```python
def read_data(file):
    with open(file, 'r') as in_file:
        lines = in_file.readlines()
        
        temps = []
        energies = []
        capacitances = []
        acceptances = []
        
        for line in lines:
            temps.append(float(line.split()[0]))
            energies.append(float(line.split()[1]))
            capacitances.append(float(line.split()[2]))
            acceptances.append(float(line.split()[3]))
        
    return temps, energies, capacitances, acceptances
```

```python
def plot_E_overlay(metropolis_file, wolff_file):
    T_m, E_m, C_m, A_m = read_data(metropolis_file)
    T_w, E_w, C_w, A_w = read_data(wolff_file)

    plt.plot(T_m, E_m, linestyle='-', marker='^', markersize=8, color='green', label='Metropolis')
    plt.plot(T_w, E_w, linestyle='None', marker='o', markersize=8, color='gold', label='Wolff')
    
    plt.xlabel(r'$T$')
    plt.ylabel(r'$E / N$', rotation=0, labelpad=40)
    
    plt.legend()
    
    plt.show()
```

```python
def plot_C_overlay(metropolis_file, wolff_file):
    T_m, E_m, C_m, A_m = read_data(metropolis_file)
    T_w, E_w, C_w, A_w = read_data(wolff_file)

    plt.plot(T_m, C_m, linestyle='-', marker='^', markersize=8, color='green', label='Metropolis')
    plt.plot(T_w, C_w, linestyle='None', marker='o', markersize=8, color='gold', label='Wolff')
    
    plt.xlabel(r'$T$')
    plt.ylabel(r'$C / N$', rotation=0, labelpad=40)
    
    plt.legend()
    
    plt.show()
```

```python
def plot_acceptance_rate(metropolis_file):
    T_m, E_m, C_m, A_m = read_data(metropolis_file)
    plt.plot(T_m, A_m, linestyle='None', marker='^', markersize=8, color='green', label='Metropolis')

    plt.xlabel(r'$T$')
    plt.ylabel('Acceptance Rate', labelpad=30)
    
    plt.legend()
    
    plt.show()
```

```python
def plot_cluster_size(wolff_file):
    T_w, E_w, C_w, A_w = read_data(wolff_file)
    plt.plot(T_w, A_w, linestyle='None', marker='o', markersize=8, color='gold', label='Wolff')

    plt.xlabel(r'$T$')
    plt.ylabel('Cluster Size', labelpad=30)
    
    plt.legend()
    
    plt.show()
```

```python
plot_E_overlay("metropolis_data.txt", "wolff_data.txt")
```

```python
plot_C_overlay("metropolis_data.txt", "wolff_data.txt")
```

```python
plot_acceptance_rate('metropolis_data.txt')
```

```python
plot_cluster_size('wolff_data.txt')
```
